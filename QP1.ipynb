```python
#Q1. Euclidean Distance Between Two Pandas Series

import pandas as pd
import numpy as np

def euclidean_distance(series1, series2):
    array1 = series1.to_numpy()
    array2 = series2.to_numpy()
    distance = np.sqrt(np.sum((array1 - array2) ** 2))
    return distance

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])
print(euclidean_distance(s1, s2))
```

```python
#Q2. Find Positions Neighbored by Smaller Values

def find_positions(series):
    result = []
    for i in range(1, len(series) - 1):
        if series[i] > series[i - 1] and series[i] > series[i + 1]:
            result.append(i)
    return result

s = pd.Series([1, 3, 7, 1, 2, 6, 3])
print(find_positions(s))
```

```python
#Q3. Replace Missing Spaces with Least Frequent Character

def replace_missing_spaces(string):
    char_count = pd.Series(list(string.replace(" ", ""))).value_counts()
    least_freq_char = char_count.idxmin()
    return string.replace(" ", least_freq_char)

input_string = "a b c d a b c"
print(replace_missing_spaces(input_string))
```

```python
#Q4. Compute Autocorrelation for a Given Lag
import pandas as pd

def compute_autocorrelations(series, lag):
    return series.autocorr(lag)

s = pd.Series([1, 2, 7, 3, 6, 5])
print(compute_autocorrelations(s, 1))
```

```python
#Q5. Find All Sundays of a Year

def sundays_of_year(year):
    all_dates = pd.date_range(start=f'{year}-01-01', end=f'{year}-12-31')
    sundays = all_dates[all_dates.dayofweek == 6]
    return sundays

print(sundays_of_year(2024))
```

```python
#q6. Convert Series to DataFrame

def convert_series_to_dataframe(series):
    return series.reset_index()

s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
print(convert_series_to_dataframe(s))
```

```python
#Q7. Stack Series Vertically and Horizontally

def stack_series(series1, series2):
    vertical = pd.concat([series1, series2], axis=0)
    horizontal = pd.concat([series1, series2], axis=1)
    return vertical, horizontal

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])
vertical, horizontal = stack_series(s1, s2)
print("Vertical:\n", vertical)
print("Horizontal:\n", horizontal)
```

```python
#Q8. Pearsons Coefficient 

def pearsons_coefficient(x, y):
    x = np.array(x)
    y = np.array(y)
    rxy = np.corrcoef(x, y)[0, 1]
    return rxy

x = [6, 8, 10]
y = [12, 10, 20]
print(pearsons_coefficient(x, y))
```

```python
#Q9. Spearman Rank Correlation(package scipy)

from scipy.stats import spearmanr

def spearman_rank_correlation(x, y):
    correlation, _ = spearmanr(x, y)
    return correlation

study_hours = [2, 4, 6, 8, 10]
exam_scores = [50, 55, 65, 75, 85]
print(spearman_rank_correlation(study_hours, exam_scores))
```

```python
#Q10. Python program to compare the two lists of customer names using the FuzzyWuzzy(package - fuzzywuzzy[python-levenshtein])

from fuzzywuzzy import fuzz
from fuzzywuzzy import process

list1 = ['John Doe', 'Jane Smith', 'Mary Jane', 'William Brown']
list2 = ['Jon Doe', 'Jane Smtih', 'Marie Jane', 'William Browne']

def compare_names(list1, list2):
    matches = []
    for name1 in list1:
        best_match = process.extractOne(name1, list2, scorer=fuzz.ratio)
        matches.append((name1, best_match[0], best_match[1]))  
    return matches

results = compare_names(list1, list2)

for name1, name2, score in results:
    print(f"{name1} matches with {name2} with a similarity score of {score}")
```
